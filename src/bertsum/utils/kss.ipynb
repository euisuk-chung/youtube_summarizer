```python
import kss
import json
import re
```

```python
kbs_script_pth = '/home/sks/korea_univ/21_1/TA/project/video_dataset/youtube-summarization/data/label/KBS뉴스_3dHyHy-lq_8_27m_55s.txt'

with open(kbs_script_pth, 'rb') as rr:
    script_set = json.load(rr)
```

```python
script = script_set['text']
```

```python
# 소수점 제외한 온점은 모두 제거
script_nopoint = re.sub('([^\d])(\.)', lambda m: "{}".format(m.group(1)), script)
```

```python
result = '. '.join(kss.split_sentences(script_nopoint))
```

```python
re.sub('(니다)([^\.])', lambda m: "{}. ".format(m.group(1)), result)
```
